```python
# Imports
import mtcnn
import cv2
import os
from pathlib import Path
```

```python
def gaussianROI( region ):
    '''
        Blur a region of an image by Gaussian blur.
    '''
    height, width, _ = region.shape
    return cv2.GaussianBlur( region, ( 11, 11 ), 5 )
```

```python
# Strong = 4
# Medium = 12
# Weak = >32
pixelateLevel = 4

pixelatedHeight, pixelatedWidth = ( pixelateLevel, pixelateLevel )

def pixelateROI( region ):
    '''
        Pixelate a region of an image by downsizing then resizing back to the original.
    '''
    height, width, _ = region.shape
    
    # Will resize the region of interest down to 32 x 32, or 12 x 12, etc etc
    resized = cv2.resize( region, ( pixelatedWidth, pixelatedHeight ), interpolation = cv2.INTER_LINEAR )
    
    # Will resize the 32 x 32 ROI back up to the original width and height
    return cv2.resize( resized, ( width, height ), interpolation = cv2.INTER_NEAREST )
```

```python
_dir = "/home/devin/Downloads/WIDER_train/images"
extension = "*.jpg"
edge = 5

stop_after = 5

print( "Loading in MTCNN weights." )
detector = mtcnn.MTCNN( './data/mtcnn_weights.npy' )

print( "Loading file list from {0}.".format( _dir ) )
files = [ str(path) for path in Path( _dir ).rglob( extension ) ]

current_file = 0
num_files = len(files)
print( "Loaded {0} files.".format( num_files ) )

total_faces = 0
g_faces = 0
p_faces = 0

for filename in files:
    try: 
        img_temp = cv2.imread( filename )
        if img_temp is None:
            print( "Skipping image {0}.".format( current_file ) )
            current_file += 1
            continue

        # Load separate images for the gaussian blur and the pixelated blur
        img_g = cv2.cvtColor( img_temp , cv2.COLOR_BGR2RGB )
        img_p = cv2.cvtColor( img_temp, cv2.COLOR_BGR2RGB )

        if img_g is None or img_p is None:
            print( "Skipping image {0}.".format( current_file ) )
            current_file += 1
            continue

        # Detect the faces in the original photo. Only done once since both
        # pictures are the same
        faces = detector.detect_faces( img_g )
        num_faces = len( faces )
        total_faces += num_faces

        # Loop through all of the faces and blur them
        for idx in range(num_faces):
            box = faces[idx]['box']
            x, y, w, h = box

            # Get the ROIs for each type of blur
            modified_roi_g = gaussianROI( img_g[ y - edge:y + h + edge, x - edge:x + w + edge ] )
            modified_roi_p = pixelateROI( img_p[ y - edge:y + h + edge, x - edge:x + w + edge ] )

            # Modify each image
            img_g[ y - edge:y + h + edge, x - edge:x + w + edge ] = modified_roi_g
            img_p[ y - edge:y + h + edge, x - edge:x + w + edge ] = modified_roi_p


        gaussian_faces  = detector.detect_faces( img_g )
        pixelated_faces = detector.detect_faces( img_p )

        gaussian_faces_found = len( gaussian_faces )
        pixelated_faces_found = len( pixelated_faces )

        g_faces += gaussian_faces_found
        p_faces += pixelated_faces_found
        current_file += 1

        print( "Done with file {0} out of {1}".format( current_file, num_files ) )
    except:
        print( "Skipping image {0}.".format( current_file ) )
        current_file += 1
    
    # stop_after -= 1
    # if stop_after < 0:
    #    print( "Stopping." )
    #    break
        
gaussian_percentage = round(float( g_faces / total_faces ) * 100, 2)
pixelated_percentage = round(float( p_faces / total_faces ) * 100, 2)

print("{0} total faces recognized in {1} photos.".format( total_faces, num_files ))
print("{0} of {1} ({2}%) faces recognized after applying gaussian blur.".format(g_faces, total_faces, gaussian_percentage))
print("{0} of {1} ({2}%) faces recognized after applying pixelated blur.".format(p_faces, total_faces, pixelated_percentage))
    
    
```
